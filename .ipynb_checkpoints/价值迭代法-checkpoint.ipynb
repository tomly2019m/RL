### 价值迭代法
* 动作价值：简单来说，就是在某一个状态下，采取特定动作后所能够获得的长期累积回报的期望值。
* 状态价值：在某一个状态下，从该状态开始采取一系列动作所能够获得的长期累积回报的期望值。

两者之间的区别：
* 动作价值强调的是动作带来的价值，而状态价值强调的是处于某种状态所带来的价值。一个更加关系下一步该采取什么动作，而另外一个更加关系下一步进入哪种状态更好。

### 贝尔曼方程

$$
V^{\pi}(s)=\max_{a} \mathbb{E}[R_{s,a}+\gamma * V^{\pi}(s(s,a))]
$$

贝尔曼方程描述的是状态价值函数，方程左边表示状态$s$的状态价值$V$,方程的右边$R_{s,a}$代表在状态$s$下采取动作$a$的奖励，而后半部分代表着对下一步状态价值的折扣。

基于马尔科夫决策过程（下一个状态仅由上一个状态所决定）

初始化$\theta_0$

```python
theta_0 = np.array([[np.nan, 1, 1, np.nan],  # s0
                    [np.nan, 1, np.nan, 1],  # s1
                    [np.nan, np.nan, 1, 1],  # s2
                    [1, 1, 1, np.nan],  # s3
                    [np.nan, np.nan, 1, 1],  # s4
                    [1, np.nan, np.nan, np.nan],  # s5
                    [1, np.nan, np.nan, np.nan],  # s6
                    [1, 1, np.nan, np.nan],  # s7、※s8はゴールなので、方策はなし
                    ])
```

```python
a, b = theta_0.shape
```

乘以$\theta_0$的目的是Q的墙壁方向值保持为nan

```python
Q = np.random.rand(a, b) * theta_0
```

```python
def simple_convert_into_pi_from_theta(theta):
    '''単純に割合を計算する'''

    [m, n] = theta.shape  # thetaの行列サイズを取得
    pi = np.zeros((m, n))
    for i in range(0, m):
        pi[i, :] = theta[i, :] / np.nansum(theta[i, :])  # 割合の計算

    pi = np.nan_to_num(pi)  # nanを0に変換

    return pi
```

```python
pi_0 = simple_convert_into_pi_from_theta(theta_0)
```

$\epsilon$-贪心法
以$\epsilon$概率随机移动，$1-\epsilon$采用最大$Q$值对应的动作

```python
def get_action(s, Q, epsilon, pi_0):
    direction = ["up", "right", "down", "left"]

    # 行動を決める
    if np.random.rand() < epsilon:
        # εの確率でランダムに動く
        next_direction = np.random.choice(direction, p=pi_0[s, :])
    else:
        # Qの最大値の行動を採用する
        next_direction = direction[np.nanargmax(Q[s, :])]

    # 行動をindexに
    if next_direction == "up":
        action = 0
    elif next_direction == "right":
        action = 1
    elif next_direction == "down":
        action = 2
    elif next_direction == "left":
        action = 3

    return action
```

```python
def get_s_next(s, a, Q, epsilon, pi_0):
    direction = ["up", "right", "down", "left"]
    next_direction = direction[a]  # 行動aの方向

    # 行動から次の状態を決める
    if next_direction == "up":
        s_next = s - 3  # 上に移動するときは状態の数字が3小さくなる
    elif next_direction == "right":
        s_next = s + 1  # 右に移動するときは状態の数字が1大きくなる
    elif next_direction == "down":
        s_next = s + 3  # 下に移動するときは状態の数字が3大きくなる
    elif next_direction == "left":
        s_next = s - 1  # 左に移動するときは状態の数字が1小さくなる

    return s_next
```

Sarsa算法，使用上一步的状态、动作、和即时奖励以及下一步的状态、动作来更新动作价值函数
$$
Q(s_t,a_t)=Q(s_t,a_t) + \eta * (R_{t+1}+\gamma Q(s_{t+1}, a_{t+1})-Q(s_t,a_t))
$$

如果已经到达了目标，下一个状态是不存在的

```python
def Sarsa(s, a, r, s_next, a_next, Q, eta, gamma):

    if s_next == 8:  # ゴールした場合
        Q[s, a] = Q[s, a] + eta * (r - Q[s, a])

    else:
        Q[s, a] = Q[s, a] + eta * (r + gamma * Q[s_next, a_next] - Q[s, a])

    return Q
```

```python
def goal_maze_ret_s_a_Q(Q, epsilon, eta, gamma, pi):
    s = 0  # 初始位置是0
    a = a_next = get_action(s, Q, epsilon, pi)  # 初始动作
    s_a_history = [[0, np.nan]] # 记录状态动作序列

    while 1:
        a = a_next  # 更新动作

        s_a_history[-1][1] = a

        s_next = get_s_next(s, a, Q, epsilon, pi)
        # 获取下一个状态

        s_a_history.append([s_next, np.nan])
        # 记录下一个状态 下一个动作未知

        # 如果到达了目标，就给予奖励 否则不给奖励 并获取下一个动作
        if s_next == 8:
            r = 1
            a_next = np.nan
        else:
            r = 0
            a_next = get_action(s_next, Q, epsilon, pi)

        # 更新Q值 动作价值函数
        Q = Sarsa(s, a, r, s_next, a_next, Q, eta, gamma)

        # 结束条件
        if s_next == 8:
            break
        else:
            s = s_next

    return [s_a_history, Q]
```

```python
eta = 0.1  # 学习率
gamma = 0.9  # 时间折扣率
epsilon = 0.5  # ε-贪心
v = np.nanmax(Q, axis=1)  # 状态价值的初始值，每次都选择动作价值最大的作为状态价值
is_continue = True
episode = 1

while is_continue:
    print("当前轮次:" + str(episode))

    # ε每次减半
    epsilon = epsilon / 2

    # 用Sarsa算法更新Q值（动作价值函数Q）
    [s_a_history, Q] = goal_maze_ret_s_a_Q(Q, epsilon, eta, gamma, pi_0)

    # 状态价值
    new_v = np.nanmax(Q, axis=1)
    print(np.sum(np.abs(new_v - v)))  # 记录状态价值的变化
    v = new_v

    print("求解迷宫问题的步数" + str(len(s_a_history) - 1))

    # 重复循环100次
    episode = episode + 1
    if episode > 1000:
        break
```

```python
from matplotlib import animation
from IPython.display import HTML

def init():
    '''背景画像の初期化'''
    line.set_data([], [])
    return (line,)

def animate(i):
    '''フレームごとの描画内容'''
    state = s_a_history[i][0]  # 現在の場所を描く
    x = (state % 3) + 0.5  # 状態のx座標は、3で割った余り+0.5
    y = 2.5 - int(state / 3)  # y座標は3で割った商を2.5から引く
    line.set_data(x, y)
    return (line,)

anim = animation.FuncAnimation(fig, animate, init_func=init, frames=len(
    s_a_history), interval=200, repeat=False)

HTML(anim.to_jshtml())
```

### Q学习算法

与Sarsa算法类似，Q学习算法的更新公式如下：
$$
Q(s_t, a_t)=Q(s_t,a_t)+\eta*(R_{t+1}+\gamma\max_aQ(s_{t+1},a)-Q(s_t,a_t))
$$

相比较于Sarsa算法，Q学习算法采用下一步状态中所有动作中动作价值最大的那个作为更新项进行更新。

```python
def Q_learning(s, a, r, s_next, Q, eta, gamma):

    if s_next == 8:  
        Q[s, a] = Q[s, a] + eta * (r - Q[s, a])

    else:
        Q[s, a] = Q[s, a] + eta * (r + gamma * np.nanmax(Q[s_next,: ]) - Q[s, a])

    return Q
```

```python
def goal_maze_ret_s_a_Q(Q, epsilon, eta, gamma, pi):
    s = 0  # 初始位置是0
    a = a_next = get_action(s, Q, epsilon, pi)  # 初始动作
    s_a_history = [[0, np.nan]] # 记录状态动作序列

    while 1:
        a = a_next  # 更新动作

        s_a_history[-1][1] = a

        s_next = get_s_next(s, a, Q, epsilon, pi)
        # 获取下一个状态

        s_a_history.append([s_next, np.nan])
        # 记录下一个状态 下一个动作未知

        # 如果到达了目标，就给予奖励 否则不给奖励 并获取下一个动作
        if s_next == 8:
            r = 1
            a_next = np.nan
        else:
            r = 0
            a_next = get_action(s_next, Q, epsilon, pi)

        # 更新Q值 动作价值函数
        Q = Q_learning(s, a, r, s_next, Q, eta, gamma)

        # 结束条件
        if s_next == 8:
            break
        else:
            s = s_next

    return [s_a_history, Q]
```

```python
a, b = theta_0.shape
Q = np.random.rand(a, b) * theta_0 * 0.1
```

```python
eta = 0.1  # 学习率
gamma = 0.9  # 时间折扣率
epsilon = 0.5  # ε-greedy
v = np.nanmax(Q, axis=1)  # 初始状态价值
is_continue = True
episode = 1

V = []  # 状态价值
V.append(np.nanmax(Q, axis=1))

while is_continue:
    print("回合数" + str(episode))

    epsilon = epsilon / 2

    [s_a_history, Q] = goal_maze_ret_s_a_Q(Q, epsilon, eta, gamma, pi_0)

    new_v = np.nanmax(Q, axis=1) 
    print(np.sum(np.abs(new_v - v))) 
    v = new_v
    V.append(v) 

    print("步数" + str(len(s_a_history) - 1))

    episode = episode + 1
    if episode > 100:
        break
```

```python
anim = animation.FuncAnimation(fig, animate, init_func=init, frames=len(
    s_a_history), interval=200, repeat=False)

HTML(anim.to_jshtml())
```
